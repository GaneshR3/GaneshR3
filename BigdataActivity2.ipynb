# Step 3: Handling Missing Values
I checked for missing values using the isnull() function:

```python
import pandas as pd

# Load the dataset, specifying the delimiter as semicolon
df = pd.read_csv("bank-full[1].csv", delimiter=";")

# Display the first few rows
df.head()

# Check for missing values
print(df.isnull().sum())

# Fill missing values with the mean for numeric columns only
for col in df.select_dtypes(include=['number']): # loop through columns with numeric data type
    df[col].fillna(df[col].mean(), inplace=True)  # Fill missing values in that column with the column's mean
```

above the steps handles the missing values by filling them with the mean for numerical columns.

# Task 2: Preprocessing an Unstructured Dataset (Text)
I opened and read a .txt file using Pythonâs built-in open() function:




```python
# Step 1: Open and Read the Text File
with open("BigData.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Print the content of the file
print(text)
```

# Step 2: Tokenization
used the Natural Language toolkit(NLTK) to tokenize the text into smaller units words

```python
import nltk
from nltk.tokenize import word_tokenize

# Download the 'punkt_tab' data for sentence tokenization
nltk.download('punkt_tab')

# Download the 'punkt' data for word tokenization
nltk.download('punkt')

tokens = word_tokenize(text)
print(tokens)
```

# Step 3: Remove Stopwords and Punctuation
Stopwords (common words like "the", "is", "and") and punctuation were removed to clean the text:


```python
from nltk.corpus import stopwords
import string
nltk.download('stopwords')

stop_words = set(stopwords.words("english"))
cleaned_text = [word for word in tokens if word.lower() not in stop_words and word not in string.punctuation]

print(cleaned_text)
```
